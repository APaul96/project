### Notes about the 'tag-list'

The tag list is a list of genres which (while the feature was still active) users could request a new genre to be added to. These genres could then be upvoted and downvoted by the community, resulting in the respective 'count' variable.
In this case we will be using the genres with the highest "ranking" in the community vote. Evidently, the community was very sparse, however this should not result in an incorrect attribution of the genre, since throughout most artists I checked, the more specific the genre got, the less votes it received. 

```python
def get_genre(artist):
    '''collect artist data from the musicbrainz database, filter out the genre information and return the genre with the highest count'''
    result = musicbrainzngs.search_artists(artist=artist)['artist-list'][0:4]
    try:
        dict_with_genre = [dictionary for dictionary in result if 'tag-list' in dictionary][0]['tag-list']
        artist_genre = ''
        genre_count = 0
        for dictionairy in dict_with_genre:
            if int(dictionairy['count']) > genre_count:
                genre_count = int(dictionairy['count'])
                artist_genre = dictionairy['name']
    except IndexError:
        artist_genre = 'Unknown'
    return artist_genre
```

```python
songdata = pd.read_csv('songdata.csv')
frame=pd.read_csv('songdata.csv')
for artist in songdata.artist.unique()[637:]:
    if os.path.isfile(artist + ':csv'):
        continue
    else:
        frame['artist']== artist
        frame['genre'] = get_genre(artist)
        frame2 = frame[frame['artist']== artist]
        frame3 = frame2[frame2['genre'] == get_genre(artist)]
        frame3.to_csv(artist + '.csv', encoding='utf-8', index=False)

###PROBLEM: Had to be started multiple times with different indexes of songdata.artist.unique due to TimeoutErrors.
### Each time I started the loop from the place in the artist.unique() list where it had stopped the time before.
```

```python
genre_count = {}
for artist in songdata.artist.unique():
    filename = artist + ".csv"
    artist_file = pd.read_csv(filename)
    if artist_file.genre.unique()[0] in genre_count:
        genre_count[artist_file.genre.unique()[0]] += 1
    else:
        genre_count[artist_file.genre.unique()[0]] = 1
    
```

```python
def outsource_genre_lyrics():
    '''filter out lyrics from all artists and export them to genre text files with a maximum number of 25 artists per genre'''
    genre_counter = {}
    for artist in songdata.artist.unique():
        file_name = artist + '.csv'
        artist_file = pd.read_csv(file_name)
        genre = artist_file.genre.unique()[0]
        if genre in genre_count:
            if genre not in genre_counter:
                genre_counter[genre] = 1
            if genre_counter[genre] <= 25:
                genre_counter[genre] += 1
                genre_csv = artist_file[['text']].copy()
                genre_csv.to_csv(genre + '.txt', header=None, index=None, sep=' ', mode='a')
    for genre in genre_counter:
        genre_counter[genre] -= 1
    return genre_counter    
```

```python
songs_in_genres = 0
for artist in songdata.artist.unique():
    file_name = artist + '.csv'
    artist_file = pd.read_csv(file_name)
    genre = artist_file.genre.unique()[0]
    if genre in genre_count:
        songs_in_genres = songs_in_genres + len(artist_file)
```
